```python
import json

#Function for pretty printing the JSON data
def pp_json(json_thing, sort=True, indents=4):
    if type(json_thing) is str:
        print(json.dumps(json.loads(json_thing), sort_keys=sort, indent=indents))
    else:
        print(json.dumps(json_thing, sort_keys=sort, indent=indents))
    return None
```

# Displaying the required fields in a tabular format using Pandas

```python
import pandas as pd
pd.set_option('display.max_colwidth', -1)

# Extracting necessary fields from the tweets data
dict_ = {'text': [], 'created_at': [], 'favorite_count': [], 'retweet_count': [], 'no_of_images': []}  
for tweet in tweets:  
    try:
        dict_['text'].append(tweet['retweeted_status']['full_text'])
    
    except:
        dict_['text'].append(tweet['full_text'])
    dict_['created_at'].append(tweet['created_at'])
    try:
        dict_['favorite_count'].append(tweet['retweeted_status']['favorite_count'])
    
    except:
        dict_['favorite_count'].append(tweet['favorite_count'])    
    
    dict_['retweet_count'].append(tweet['retweet_count'])
    
    try:
        dict_['no_of_images'].append(len(tweet['extended_entities']['media']))
    
    except:
        dict_['no_of_images'].append('None')    
    

# Structure data in a pandas DataFrame for easier manipulation
df = pd.DataFrame(dict_)  
#Displaying the data in a tabular form￼
df
```
